```python
import pandas as pd
```

```python
df = pd.read_csv('hw.txt', header = None)
df
```

```python
dic = {}
l = df[0].tolist()
h=[]
w=[]


for i,j in enumerate(l):
    
    if i%2==0:
        h.append(j)
    else:
        w.append(j)
        

dic["Person Number"] = list(range(1,int(len(l)/2)+1))
dic["Height"] = h
dic["weight"] = w
dic["Height"] = [(i/100) for i in dic['Height']]
dic["weight"] = [(i/2.2) for i in dic['weight']]
```

```python
df = pd.DataFrame.from_dict(dic)
df['BMI'] = df['weight']/df['Height']
```

```python
categories = []
for i in df['BMI']:
    if i < 18.5:
        categories.append('Under Weight')
    elif i >= 18.5 and i <= 24.9:
        categories.append('Normal')
    elif i >= 25.9 and i <= 29.9:
        
        categories.append('Over Weight')
    else:
        
        categories.append('Obesity')
```

```python
df['Category'] = categories
df
```

```python
df1 = df[df['Category'] == 'Over Weight']
df1.reset_index(drop=True, inplace=True)

df2 = df[df['Category'] == 'Normal']
df2.reset_index(drop=True, inplace=True)

df3 = df[df['Category'] == 'Under Weight']
df3.reset_index(drop=True, inplace=True)

df4 = df[df['Category'] == 'Obesity']
df4.reset_index(drop=True, inplace=True)

```

```python
# Overweight

n = len(df1)
k=0
print('over weight')
if n == 0:
    print("None")
else:
    while k < n:  

        print("Person-",df1["Person Number"][k],":",df1["BMI"][k])

        k+=1
print(f"Group size: {n}\nAverage BMI for this group: {df1.BMI.mean()}")
#Underweight

n = len(df3)
k=0
print('under weight')
if n == 0:
    print("None")
else:
    while k < n:  

        print("Person-",df3["Person Number"][k],":",df3["BMI"][k])

        k+=1
print(f"Group size: {n}\nAverage BMI for this group: {df3.BMI.mean()}")
    
        
# Normal

n = len(df2)
k=0
print('Normal')
if n == 0:
    print("None")
else:
    while k < n:  

        print("Person-",df2["Person Number"][k],":",df2["BMI"][k])

        k+=1
print(f"Group size: {n}\nAverage BMI for this group: {df2.BMI.mean()}")
# Obese


n = len(df4)
k=0
print('Obesity')
if n == 0:
    print("None")
else:
    while k < n:  

        print("Person-",df4["Person Number"][k],":",df4["BMI"][k])

        k+=1

print(f"Group size: {n}\nAverage BMI for this group: {df4.BMI.mean()}")
    
print(f"The File has records for {len(df)} persons")
print(f"Average BMI for all is : {df.BMI.mean()}")
```
